```python
import joblib
import pandas as pd
```

```python
model = joblib.load('fertilizer_model.joblib')
model
```

```python
sample = pd.DataFrame([{
    "plant_type": "Flower Plant",
    "location": "Rooftop",
    "sunlight": "High",
    "soil_type": "Loamy",
    "soil_moisture": "Medium",
    "nitrogen": 30,
    "phosphorus": 25,
    "potassium": 40,
    "soil_ph": 6.5
}])

probs = model.predict_proba(sample)

classes = model.classes_

top5 = probs[0].argsort()[-5:][::-1]

print("Recommended Fertilizers:\n")

for i in top5:
    print(f"{classes[i]}  ({probs[0][i]*100:.2f}%)")
```
